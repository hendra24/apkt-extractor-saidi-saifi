```python
import pandas as pd
import os
from bs4 import BeautifulSoup
import math

# Get the current working directory
dir_path = os.getcwd()

##source == for data from eis apkt, processed == for finished processing file
source_dir_path = os.path.join(dir_path, "source")
processed_dir_path = os.path.join(dir_path, "processed")

#File name penampung data hasil pengolahn/ output file
file_name = "apkt_eis_saidi_saifi_combined.csv"

# Get the list of files in the directory
files = os.listdir(source_dir_path)


print("working directory : " + source_dir_path)

if len(files) != 0 :   
    print("jumlah file yang akan di prosess : " + str(len(files)))
    for file in files :
        print("processing file : " + file)
        
        #file directory
        file_path = os.path.join(source_dir_path,file)
            
        #open saved file
        with open(file_path, 'r') as f:
            soup = BeautifulSoup(f, 'html.parser')

        #CHECK TIPE FILE
        df_list = pd.read_html(file_path)
        
        try:
            math.isnan(df_list[1][1][2])
            #jika uid proses pakai metode uid
            print("uid")
            parser_data("uid")
        except:
            #jika up3 proses pkaia metode up3
            print("up3")
            parser_data("up3")
        
        
        # closed file
        f.close()
        
        print(file)
        
        #move file to processed folder
        os.rename(file_path, os.path.join(processed_dir_path, file))
        
else:
    print("tidak ada data yang akan di prosess di folder source")
        
```

```python
def parser_data(x):
    #kolum data extraksi
    kolum = ["no kode (a)", "penyebab gangguan (b)", "jml plg padam (c)", "jam x jml plg padam (d)", "saidi jam (e=d/plg)", "saifi jam (f=c/plg)", "jml ggn kali (h)", "lama padam dalam jam (i)", "kwh tak salur (j)"]
    #Ubah format tanggal dari indonesia ke english
    format_eng = {"Januari": "January", "Februari" : "February", "Maret": "March", "April": "April","Mei":"May", "Juni" : "June", "Juli": "July", "Agustus": "August", "Oktober": "October", "September":"September","Nopember": "November", "Desember":"December" }
    
    
    #extract meta data from file
    info_table = soup.find('table')
    # Extract the data from the table and store it in a list of lists
    meta_data = []
    for row in info_table.find_all('tr'):
        cols = row.find_all('td')
        cols = [col.text.strip() for col in cols]
        meta_data.append(cols)

    #print meta data 
    tanggal = meta_data[7][0]
    
    jumlah_pelanggan = meta_data[10][2]
    
            
               
    # Find the specific table in the HTML using its class or id
    table = soup.find('table', class_="gridtable")

    # Extract the data from the table and store it in a list of lists
    data = []
    for row in table.find_all('tr'):
        cols = row.find_all('td')
        cols = [col.text.strip() for col in cols]
        data.append(cols)


    df = pd.DataFrame(data[2:], columns=kolum)

    #print(df.head())
    df["jumlah pelanggan"] = jumlah_pelanggan
    #Exctract date
    tmp = str.split(tanggal)

    #Ubah format tanggal dari indonesia ke english
    tanggal = format_eng[tmp[0]]+"/"+tmp[1] 
    print(tanggal)
    
    # replace commas with dots and remove dots
    df['jml plg padam (c)'] = df['jml plg padam (c)'].str.replace('.', '').str.replace(',', '.', regex=False)
    df['jam x jml plg padam (d)'] = df['jam x jml plg padam (d)'].str.replace('.', '').str.replace(',', '.', regex=False)
    df['saidi jam (e=d/plg)'] = df['saidi jam (e=d/plg)'].str.replace('.', '').str.replace(',', '.', regex=False)
    df['saifi jam (f=c/plg)'] = df['saifi jam (f=c/plg)'].str.replace('.', '').str.replace(',', '.', regex=False)
    df['jml ggn kali (h))'] = df['jml ggn kali (h)'].str.replace('.', '').str.replace(',', '.', regex=False)
    df['lama padam dalam jam (i)'] = df['lama padam dalam jam (i)'].str.replace('.', '').str.replace(',', '.', regex=False)
    df['kwh tak salur (j)'] = df['kwh tak salur (j)'].str.replace('.', '').str.replace(',', '.', regex=False)
    df['jumlah pelanggan'] = df['jumlah pelanggan'].str.replace('.', '').str.replace(',', '.', regex=False)
    # convert the data type to float
    df['jml plg padam (c)'] = pd.to_numeric(df['jml plg padam (c)'], errors='coerce')
    df['jam x jml plg padam (d)'] = pd.to_numeric(df['jam x jml plg padam (d)'], errors='coerce')
    df['saidi jam (e=d/plg)'] = pd.to_numeric(df['saidi jam (e=d/plg)'], errors='coerce')
    df['saifi jam (f=c/plg)'] = pd.to_numeric(df['saifi jam (f=c/plg)'], errors='coerce')
    df['jml ggn kali (h))'] = pd.to_numeric(df['jml ggn kali (h))'], errors='coerce')
    df['lama padam dalam jam (i)'] = pd.to_numeric(df['lama padam dalam jam (i)'], errors='coerce')
    df['kwh tak salur (j)'] = pd.to_numeric(df['kwh tak salur (j)'], errors='coerce')
    df['jumlah pelanggan'] =pd.to_numeric(df['jumlah pelanggan'], errors='coerce')
    
    if x == 'uid' :
        
        #Add date and unit column
        df["unit"] = "UID"
        df["tanggal"] = tanggal
        
        
        
        #process data by category
        process_data_unplaned("uid",df)
        process_data_planed("uid",df)
        process_data_bencana("uid",df)
        
        
    else:
        
        #extract unit
        unit = df_list[1][1][2][6:]
        print(unit)
        
        #Add date and unit column
        df["unit"] = unit
        df["tanggal"] = tanggal
        
        #print(df.head())
        
       #process data by category
        process_data_unplaned("up3",df)
        process_data_planed("up3",df)
        process_data_bencana("up3",df)
        
        
```
